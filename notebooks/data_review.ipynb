```python
import pandas as pd
import os
```

```python
DATA_DIR = os.path.join("..", "data")
RAW_DIR = os.path.join(DATA_DIR, "raw")

AUTHORS_PATH = os.path.join(DATA_DIR, "authors.csv")
RECORDS_PATH = os.path.join(DATA_DIR, "records.csv")
```

```python
authors_df = pd.read_csv(AUTHORS_PATH, on_bad_lines = "warn")
authors_df
```

```python
records_df = pd.read_csv(RECORDS_PATH, on_bad_lines = "warn")
records_df
```

```python
# see if any books have missing titles
missing_titles = records_df[records_df["title"].isna()]
missing_titles
```

```python
# removing missing titles
records_df.dropna(inplace = True)
records_df.reset_index(inplace = True)
```

```python
records_df["title"].str.extract(r"\(([^)]*)\)").dropna()[0].unique()
```

```python
def read_book_raw(name):
    path = os.path.join(RAW_DIR, name)
    with open(path, "r") as f:
        return f.read()

record_paths = records_df["author_id"].astype(str) + "_" + records_df["id"].astype(str) + ".txt"

records_df["text"] = record_paths.apply(read_book_raw)
records_df
```
